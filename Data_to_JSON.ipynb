```python
import pandas as pd
import numpy as np
```

### Fetch data from here: https://www.basketball-reference.com/leagues/NBA_2020_advanced.html

```python
getAdvancedStats = pd.read_csv("2019-2020.csv")
getAdvancedStats
```

```python
AdvancedStats = getAdvancedStats[['Player', 'PER', 'VORP', 'WS', 'MP']]
AdvancedStats
```

```python
## NEED TO REMOVE PLAYERS THAT MOVED TEAMS TO ONLY KEEP THEIR TOTAL ROW

without_dups = AdvancedStats[~AdvancedStats["Player"].duplicated(keep='first')]
without_dups
```

```python
neg_vorp = without_dups[without_dups['VORP']<0]
neg_vorp
```

```python
neg_WS = without_dups[without_dups['WS']<0]
neg_WS
```

```python
def determine_multiplication(series_1, series_2):
    if series_1 < 0 and series_2 < 0:
        return - series_1*series_2
    else:
        return series_1 * series_2

without_dups['WS_VORP'] = without_dups.apply(lambda row: (row['VORP']*row['WS']) if row['VORP'] >= 0 and row['WS'] >=0 else -(row['VORP']*row['WS']), axis=1)
```

```python
without_dups['Utility'] = without_dups['PER'] + without_dups['WS_VORP']
```

```python
without_dups.sort_values('Utility', ascending=False)
```

```python
final = without_dups.set_index('Player')[['PER', 'VORP', 'WS', 'Utility']]
final
```

```python
final.to_json('player_data.json',orient='index')
```
